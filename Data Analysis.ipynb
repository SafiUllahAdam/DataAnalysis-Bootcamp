**Drop, FIll, Save New Updated Data**

**EMpty Values (NaN) have'nt been deleted yet ,  SO 1st drop these values with Dropna**

```python
# Fill missing data with the mean Score
# column_name.fillna(which_column_want_to_replace)

df['iq'].fillna(df['iq'].mean(), inplace=True)
print("\nDataFrame with Missing Scores Filled:")
df
```

```python
df_dropped = df.dropna()
print("\nDataFrame with Missing Values Dropped:")
df_dropped
```

```python
# Filter the DataFrame to keep only rows where iq > 116
df = df[df['iq'] > 117]

# Print the updated DataFrame
print("\nDataFrame with 'iq' greater than 116:")
df
```
