**Try & Catch Errors**

```python
my_tuple = (1, 4, 7)
print(my_tuple)
```

try: Runs the code you expect might cause an error.

except: Handles the error gracefully without breaking the program.

as e: Stores the actual error message (e.g., 'tuple' object does not support item assignment).

This is helpful to handle situations where your program might crash due to errors like trying to modify a tuple, which is not allowed.

**Classes**

**Class: Template for creating objects (students in this case).

self: Refers to the instance (specific student).

__init__(): Initializes the object with values (name, grade).

Method (show_info): Allows you to perform actions with the object, like showing info.**

```python
#Defining a simple class : Student Class

class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def show_info(self):
        return(f"Student: {self.name}, Grade: {self.grade}")

    def show_result(self):
      if self.grade >= 2.5:
        return f"Student {self.name} Has Passed "
      else :
        return f"Student {self.name} Has Failed "


student1 = Student("Safi", 2.9)
student2 = Student("Eesha", 2.95)
print(student1.show_info())
print(student2.show_info())
print(student1.show_result())
print(student2.show_result())

```

```python
class customer:
    def __init__(self,name,  email, phone):
     self.name = name
     self.email = email
     self.phone = phone

    def show_cust_details(self):
      return f"Details of Customer {self.name} is"

    def place_order(self):
        return f"Order placed for {self.name}."

    def cancel_order(self):
        return f"Order canceled for {self.name}."

# Get input for the first customer
name1 = input("Welcome: Please enter your name for the first customer: ")
email1 = input("Please enter your email for the first customer: ")
phone1 = input("Please enter your phone number for the first customer: ")
customer1 = customer(name1, email1, phone1)


# Get input for the second customer
name2 = input("Welcome: Please enter your name for the second customer: ")
email2 = input("Please enter your email for the second customer: ")
phone2 = input("Please enter your phone number for the second customer: ")
customer2 = customer(name2, email2, phone2)


# Display the details of both customers
print(customer1.show_cust_details())
print(f"Name: {customer1.name}, Email: {customer1.email}, Phone: {customer1.phone}")
print(customer1.place_order())
print(customer1.cancel_order())

print(customer2.show_cust_details())
print(f"Name: {customer2.name}, Email: {customer2.email}, Phone: {customer2.phone}")
print(customer2.place_order())
print(customer2.cancel_order())
```
